```python
import pandas as pd
import sklearn.model_selection as skms
import sklearn.linear_model as sklm
from sklearn.metrics import mean_squared_error,r2_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
paddy_yala_dataset=pd.read_csv(r'D:\Thimantha\Documents2\Data Science\paddy_production_in_yala_season_1952_2012.csv')
```

```python
paddy_yala_dataset.head(20)
```

```python
x=paddy_yala_dataset.loc[:,['Yala Season','Sown 000 Acres']]
y=paddy_yala_dataset.loc[:,['Harvested 000 Acres']]
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=skms.train_test_split(x,y,test_size=1/4,random_state=0)
```

```python
print(x_train)
```

```python
print(y_train)
```

```python
lm_yala=sklm.LinearRegression()
```

```python
lm_yala.fit(x_train,y_train)
```

```python
prediction=lm_yala.predict(x_test)
```

```python
print("Mean squared error : %.2f"
      % mean_squared_error(y_test,prediction))


print('Variance score: %.2f' % r2_score(y_test,prediction))
```

```python
plt.scatter(y_test,prediction)
plt.show()
```
